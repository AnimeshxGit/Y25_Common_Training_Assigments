```python
!pip install scipy -q

import numpy as np
from scipy.io import wavfile
from scipy.signal import correlate

sample_rate, audio = wavfile.read("/content/sample_data/Pikachu.wav")

audio = audio.astype(np.float32)

if len(audio.shape) > 1:
    audio = np.mean(audio, axis=1)

delay = sample_rate

signal0 = audio.copy()

signal1 = np.pad(audio, (delay, 0))[:-delay]

signal2 = np.pad(audio, (2 * delay, 0))[:-(2 * delay)]

signal3 = np.pad(audio, (3 * delay, 0))[:-(3 * delay)]

wavfile.write("delay_0s.wav", sample_rate, signal0.astype(np.int16))
wavfile.write("delay_1s.wav", sample_rate, signal1.astype(np.int16))
wavfile.write("delay_2s.wav", sample_rate, signal2.astype(np.int16))

ref1 = signal1

c1 = np.max(correlate(signal0, ref1, mode='same'))
c2 = np.max(correlate(signal1, ref1, mode='same'))
c3 = np.max(correlate(signal2, ref1, mode='same'))

print("Part A")
print("0s with 1s reference =", c1)
print("1s with 1s reference =", c2)
print("2s with 1s reference =", c3)

new_ref = signal0 + signal2

d1 = np.max(correlate(signal0, new_ref, mode='same'))
d2 = np.max(correlate(signal1, new_ref, mode='same'))
d3 = np.max(correlate(signal2, new_ref, mode='same'))
d4 = np.max(correlate(signal3, new_ref, mode='same'))

print("\nPart B")
print("0s with combined reference =", d1)
print("1s with combined reference =", d2)
print("2s with combined reference =", d3)
print("3s with combined reference =", d4)
```
